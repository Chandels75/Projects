```python
#92. Customer Segmentation: Segment customers based on their purchase behavior.
rfm_df=df.groupby(['City', 'Purchase Address']).agg({'Order Date':lambda x:(pd.Timestamp.now() - x.max()).days,
                                                     'Order ID': 'nunique','Total_Price':'sum',}).reset_index()
rfm_df.columns = ['City', 'Purchase Address', 'Recency', 'Frequency', 'Total_Price']
# Standardize the RFM metrics
scaler = StandardScaler()
rfm_df[['Recency', 'Frequency', 'Total_Price']] = scaler.fit_transform(rfm_df[['Recency', 'Frequency', 'Total_Price']])
```

```python
# Find the optimal number of clusters using the Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)  # Correctly define kmeans
    kmeans.fit(rfm_df[['Recency', 'Frequency', 'Total_Price']])
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method graph
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.xticks(range(1, 11))
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#100. Seasonal Analysis: Analyze sales patterns across different seasons.
def get_season(month):
    if month in [12,1,2]:
        return 'winter'
    elif month in [3,4,5]:
        return 'Spring'
    elif month in [6,7,8]:
        return 'Summer'
    else:
        return 'Fall'
df['Season']=df['Order Date'].dt.month.apply(get_season)
# Aggregate sales by season
seasonal_sales=df.groupby('Season')['Total_Price'].sum().reset_index()
# Order the seasons
seasonal_sales['Season'] = pd.Categorical(seasonal_sales['Season'], categories=['Winter', 'Spring', 'Summer', 'Fall'], ordered=True)
seasonal_sales = seasonal_sales.sort_values('Season')
# Plot the seasonal sales
plt.figure(figsize=(10, 6))
sns.barplot(x='Season', y='Total_Price', data=seasonal_sales, palette='coolwarm')
plt.title('Total Sales by Season')
plt.xlabel('Season')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()

# Additional analysis: Average sales per season
avg_seasonal_sales = df.groupby('Season')['Total_Price'].mean().reset_index()
avg_seasonal_sales['Season'] = pd.Categorical(avg_seasonal_sales['Season'], categories=['Winter', 'Spring', 'Summer', 'Fall'], ordered=True)
avg_seasonal_sales = avg_seasonal_sales.sort_values('Season')

# Plot the average sales per season
plt.figure(figsize=(10, 6))
sns.barplot(x='Season', y='Total_Price', data=avg_seasonal_sales, palette='coolwarm')
plt.title('Average Sales by Season')
plt.xlabel('Season')
plt.ylabel('Average Sales')
plt.tight_layout()
plt.show()
```
